```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

brand_ratings = pd.read_csv('http://bit.ly/PMR-ch9')
brand_ratings.head()

#Looking at the tail
brand_ratings.tail(2)

#Looking at the describe and round by 2
brand_ratings.describe().T.round(2)

#use boxplot to view the distribution
brand_ratings.plot.box()
```

```python
#data can be standardized with an expression using mean and stdv
x = np.arange(1000)
x_sc = (x - x.mean()) / x.std()
print('mean: {}\nmedian: {}\nmax: {}\nmin: {}'.format(x_sc.mean(),
                                                      np.median(x_sc),
                                                      x_sc.max(),
                                                      x_sc.min()))

#Using scale to rescale all variables at once
from sklearn.preprocessing import scale
brand_ratings_sc = brand_ratings.copy()
brand_ratings_sc.iloc[: , :-1] = scale(brand_ratings_sc.iloc[:, :-1])
brand_ratings_sc.plot.box()
brand_ratings_sc.describe().T.round(2)
```
