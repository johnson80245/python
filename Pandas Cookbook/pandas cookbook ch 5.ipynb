```python
#Comparing Categorical Values with Categorical Values

#Lower cardinality with makes and models
def generalize(ser, match_name, default):
    seen = None
    for match, name in match_name:
        mask = ser.str.contains(match)
        if seen is None:
            seen = mask
        else:
            seen |= mask
        ser = ser.where(~mask, name)
    ser = ser.where(seen, default)
    return ser

makes = ['Ford', 'Tesla', 'BMW', 'Toyota']
data = fueleco[fueleco.make.isin(makes)].assign(
    SClass = lambda df_ : generalize(
        df_.VClass,
        [
            ('Seaters', 'Car'),
            ('Car', 'Car'),
            ('Utility', 'SUV'),
            ('Truck', 'Truck'),
            ('Van', 'Van'),
            ('van', 'Van'),
            ('Wagon', 'Wagon')
        ],
        'other'
    )
)

#Summarize the counts of vehicle classes for each make
data.groupby(['make','SClass']).size().unstack()

#Use the crosstab function instead of the chain (quicker)
pd.crosstab(data.make, data.SClass)

#Add more dimensions:
pd.crosstab(
    [data.year, data.make], [data.SClass, data.VClass]
)

#Use cramers v measure 
import scipy.stats as ss

def cramers_v(x, y):
    confusion_matrix = pd.crosstab(x, y)
    chi2 = ss.chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum().sum()
    phi2 = chi2 / n
    r, k = confusion_matrix.shape
    phi2corr = max(
        0, phi2 - ((k - 1) * (r - 1)) / (n - 1)
    )
    rcorr = r - ((r - 1) ** 2) / (n - 1)
    kcorr = k - ((k - 1) ** 2) / (n - 1)
    return np.sqrt(
        phi2corr / min((kcorr - 1), (rcorr - 1))
    ) 

cramers_v(data.make, data.SClass)

#Viisuzlize the vross tab as bar plot

fig, ax = plt.subplots(figsize = (10, 8))
(
    data.pipe(
        lambda df_: pd.crosstab(df_.make, df_.SClass)
    ).plot.bar(ax = ax)
)

#Visualize the cross tab using seaborn
res = sns.catplot(kind = 'count', x = 'make', hue = 'SClass', data = data)

#Visualize the relative sizes by normalizing crosstab by stacked bar
fig, ax = plt.subplots(figsize = (10, 8))
(
    data.pipe(
        lambda df_ : pd.crosstab(df_.make, df_.SClass)
    )
    .pipe(lambda df_ : df_.div(df_.sum(axis = 1), axis = 0))
    .plot.bar(stacked = True, ax = ax)
)

```
